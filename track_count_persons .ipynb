```python
import cv2
import time
import numpy as np
from ultralytics import YOLO
from deep_sort.deep_sort import DeepSort

# Initialize video capture
video_path = "/content/20593700-uhd_2160_3840_30fps.mp4"  # Change to 0 for webcam
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()

width = int(cap.get(3))
height = int(cap.get(4))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Output video settings
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # Use XVID for better compatibility
out = cv2.VideoWriter('output.mp4', fourcc, fps, (width, height))

# Load YOLO model
model = YOLO("yolov8n.pt")

# Initialize DeepSORT tracker
deep_sort_weights = 'deep_sort/deep/checkpoint/ckpt.t7'
tracker = DeepSort(model_path=deep_sort_weights, max_age=70)

# Define vehicle classes
vehicle_classes = {2: 'car', 5: 'bus', 7: 'truck', 3: 'motorcycle'}

# Track unique IDs
unique_track_ids = set()

# FPS calculation variables
start_time = time.time()
frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("End of video or read error.")
        break

    og_frame = frame.copy()

    # YOLO Detection
    results = model(og_frame, conf=0.3, classes=list(vehicle_classes.keys()), agnostic_nms=True)

    bboxes_xywh, confidences, track_classes = [], [], []

    for result in results:
        boxes = result.boxes
        if boxes.cls is not None and len(boxes.cls) > 0:
            cls = boxes.cls.tolist()
            xywh = boxes.xywh.cpu().numpy()
            conf = boxes.conf.cpu().numpy()

            for i, class_index in enumerate(cls):
                class_index = int(class_index)
                if class_index in vehicle_classes:
                    bboxes_xywh.append(xywh[i])
                    confidences.append(conf[i])
                    track_classes.append(vehicle_classes[class_index])

    # Convert to numpy arrays
    bboxes_xywh = np.array(bboxes_xywh, dtype=float) if bboxes_xywh else np.empty((0, 4), dtype=float)
    confidences = np.array(confidences, dtype=float) if confidences else np.empty((0,), dtype=float)

    # Update tracker only if detections exist
    tracks = tracker.update(bboxes_xywh, confidences, og_frame) if len(bboxes_xywh) > 0 else []

    # Draw tracking results
    for track in tracker.tracker.tracks:
        if not track.is_confirmed() or track.time_since_update > 1:
            continue

        track_id = track.track_id
        x1, y1, x2, y2 = map(int, track.to_tlbr())

        # Assign color based on track ID
        colors = [(0, 0, 255), (255, 0, 0), (0, 255, 0)]  # Red, Blue, Green
        color = colors[track_id % len(colors)]

        # Draw bounding box and label
        cv2.rectangle(og_frame, (x1, y1), (x2, y2), color, 2)
        cv2.putText(og_frame, f"ID-{track_id}", (x1 + 10, y1 - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)

        # Track unique IDs
        unique_track_ids.add(track_id)

    # Update vehicle count
    vehicle_count = len(unique_track_ids)

    # FPS calculation
    frame_count += 1
    elapsed_time = time.time() - start_time
    fps_display = frame_count / elapsed_time if elapsed_time > 0 else 0

    # Display FPS and vehicle count
    cv2.putText(og_frame, f"Vehicles: {vehicle_count}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
    cv2.putText(og_frame, f"FPS: {fps_display:.2f}", (10, 60),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)

    # Show output frame (for debugging)
    from google.colab.patches import cv2_imshow  # Import for Colab

    # Write frame to video
    out.write(og_frame)

cap.release()
out.release()
cv2.destroyAllWindows()
print("Processing complete. Output saved as 'output.mp4'.")
```
